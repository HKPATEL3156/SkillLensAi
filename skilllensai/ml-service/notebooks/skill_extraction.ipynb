```python
# library for data handling
import pandas as pd

# library for pdf reading
import pdfplumber

# library for text processing
import re
```

```python
# library for nlp
import nltk
nltk.download('punkt')
```

```python
# sample skill list (you can later load from dataset)
skill_list = [
    "python", "java", "c++", "sql", "mysql", "mongodb",
    "react", "node", "express", "machine learning",
    "deep learning", "tensorflow", "pytorch",
    "html", "css", "javascript", "docker", "aws"
]
```

```python
# function to extract text from pdf
def extract_text_from_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + " "
    return text
```

```python
# function to clean text
def clean_text(text):
    text = text.lower()  # convert to lowercase
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)  # remove special chars
    return text
```

```python
# function to extract skills
def extract_skills(text, skill_list):
    found_skills = []
    for skill in skill_list:
        if skill.lower() in text:
            found_skills.append(skill)
    return list(set(found_skills))
```

```python
# path of resume (change this)
resume_path = "../uploads/main3.pdf"
```

```python
# extract text
raw_text = extract_text_from_pdf(resume_path)
```

```python
# clean text
cleaned_text = clean_text(raw_text)
```

```python
# get skills
skills_found = extract_skills(cleaned_text, skill_list)
```

```python
# main result
print("extracted skills:")
print(skills_found)
```
